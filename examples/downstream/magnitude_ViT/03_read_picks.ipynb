```python
import matplotlib.pyplot as plt
import numpy as np


yh2 = np.load('pred_SS_all_FoundationVit_Mag.npy', allow_pickle=True)
spt = np.load('pall_FoundationVit_MAg.npy', allow_pickle=True)
yh2_mean = np.mean(yh2,axis=1)

diff = yh2_mean-spt
mae = np.mean(np.abs(yh2_mean-spt))
std = np.std((yh2_mean-spt))
```

## Plotting

```python
font = {
        'weight' : 'bold',
        'size'   : 14}
plt.rc('font', **font)

fig = plt.figure()
ax = fig.add_subplot(1,1,1)
plt.hist(diff, bins=30)
plt.xlim([-1,1])
plt.ylim([0,15000])
plt.xlabel("Magnitude Error ",fontsize='large', fontweight='bold')
plt.ylabel("Count ",fontsize='large', fontweight='bold')
plt.grid()
ax.set_axisbelow(True)
plt.text(-0.9,2000, 'MAE=' + str(np.round(np.mean(np.abs(diff)),3)) )
plt.text(-0.9,500, '$\sigma$=' + str(np.round(np.std((diff)),3)) )
plt.show()


fig4, ax = plt.subplots()
ax.scatter(spt, yh2_mean, alpha = 0.4, facecolors='none', edgecolors='r')
ax.plot([spt.min(), spt.max()], [spt.min(), spt.max()], 'k--', alpha=0.4, lw=2)
ax.set_xlabel('Measured')
ax.set_ylabel('Predicted')
```
